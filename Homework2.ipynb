```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q1-Read the data https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv

iData = pd.read_csv('https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv', )

iData
```

```python
#Q2- Get data information and number of rows and columns
iData.info()
```

```python
#Q3-Get first 3 rows of data
iData.head(3)
```

```python
#Q4-Get descriptive statistics of data
iData.describe()
```

```python
#Q5-Use seasoborns heat map to generate following heatmap
# annot = True to print the correlation inside the square

iCorr = iData.corr()
iCorr

# this is with matplotlib library

# ifig = plt.figure(figsize=(8,8))
# plt.matshow(iCorr, cmap='RdBu', fignum=ifig.number);
# plt.xticks(range(len(iCorr.columns)), iCorr.columns, rotation='vertical');
# plt.yticks(range(len(iCorr.columns)), iCorr.columns);

#this is with seasoborns heatmap

sns.heatmap(iCorr, annot=True);

```

```python
#Q6-Plot the density distribution of Balance

iData.Balance.plot(kind='density', figsize=(10,4));
```

```python
#Q7-Count the frequency values of each Region

iData.Region.value_counts()
# iData.Region.value_counts().plot(kind='bar', figsize=(10,4), color='lightgreen');
```

```python
#Q8-List Percentage of  the Married people

iData.Married.value_counts(normalize=True).Yes

listPercentageOfMarriedPeople = ((iData.Married.value_counts().Yes / iData.Married.count()) * 100);
print(iData.Married.value_counts(normalize=True), end='\n----------------------\n')

print(iData.Married.value_counts(normalize=True).Yes, end='\n----------------------\n')

print('Percentage of Married people is ' + str(round(listPercentageOfMarriedPeople,2)) + '%')
```

```python
#Q9-Create histogram of Balance

iData.Balance.plot(kind='hist', figsize=(6,4), color='orange')
```

```python
#Q10-Get the frequency distribution of Region using bar graph

iData.Region.value_counts().plot(kind='bar', figsize=(9,4), color='lightgreen');
```

```python
#Q11-Get the mean income  of married people?

iData.groupby('Married').Income.mean().Yes
```

```python
#Q12-Get the scatter plot of Income and Credit

iData.plot(kind='scatter', x='Income', y='Limit', figsize=(7,7))
```

```python
# Q14-Create the pair plot using Seaborn library's pairplot

sns.pairplot(iData)
```
